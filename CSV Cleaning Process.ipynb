```python
# Importing Dependencies
import pandas as pd
import numpy as np
```

```python
# Uploading Original CSV files
csv1 = "JC-202203-citibike-tripdata.csv"
csv2 = "JC-202303-citibike-tripdata.csv"
citibike_data_2022 = pd.read_csv(csv1)
citibike_data_2023 = pd.read_csv(csv2)
```

```python
merged_df = pd.concat([citibike_data_2022, citibike_data_2023], ignore_index=True)
merged_df.head()
```

```python
# Convert the "started_at" and "ended_at" columns to pandas datetime objects
merged_df["started_at"] = pd.to_datetime(merged_df["started_at"])
merged_df["ended_at"] = pd.to_datetime(merged_df["ended_at"])

# Extract the date and time components into separate columns
merged_df["start_date"] = merged_df["started_at"].dt.date
merged_df["start_time"] = merged_df["started_at"].dt.time
merged_df["end_date"] = merged_df["ended_at"].dt.date
merged_df["end_time"] = merged_df["ended_at"].dt.time

# drop the original "started_at" and "ended_at" columns
merged_df.drop(columns=["started_at", "ended_at"], inplace=True)
```

```python
# Convert "start_date" and "end_date" to pandas datetime objects
merged_df["start_date"] = pd.to_datetime(merged_df["start_date"])
merged_df["end_date"] = pd.to_datetime(merged_df["end_date"])

# Convert "start_time" and "end_time" to pandas datetime objects with a specific format
merged_df["start_time"] = pd.to_datetime(merged_df["start_time"], format="%H:%M:%S").dt.time
merged_df["end_time"] = pd.to_datetime(merged_df["end_time"], format="%H:%M:%S").dt.time
```
